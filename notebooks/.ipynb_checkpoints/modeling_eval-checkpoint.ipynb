```python
import os
os.chdir('..')
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from lightgbm import LGBMClassifier
from src.preprocessing import CreditPreprocessor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/microfinance.csv')
```

```python
# Split
X = df.drop(columns=['default'])
y = df['default']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)
```

```python
cat_cols = ['employment_type', 'region', 'education', 'age_group', 'prev_loans_groups']
num_cols = [
    'age', 'income', 'prev_loans', 'credit_history_bad',
    'loan_amount', 'loan_term_days',
    'daily_payment', 'debt_to_income', 'payment_to_income',   # ← новые
]
```

```python
# Энкодеры
encoder = ColumnTransformer([
    ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), cat_cols),
    ('num', StandardScaler(), num_cols),
])
```

## Бейзлайн Лог регрессия

```python
# Пайплайн

pipeline = Pipeline([
    ('preproc', CreditPreprocessor()),
    ('encode', encoder),
    ('model', LogisticRegression(class_weight='balanced', random_state=42))
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, roc_auc_score, average_precision_score

# Вероятности дефолта на тесте
y_proba = pipeline.predict_proba(X_test)[:, 1]
y_pred = pipeline.predict(X_test)

# Основная метрика
print("PR-AUC:", average_precision_score(y_test, y_proba))
print("ROC-AUC:", roc_auc_score(y_test, y_proba))
print(classification_report(y_test, y_pred))  # recall/precision по дефолту
```

## Подбор гиперпараметров с бустингом

```python
import optuna
from sklearn.model_selection import cross_val_score, StratifiedKFold

def objective(trial):
    params = {
        'n_estimators':     trial.suggest_int('n_estimators', 100, 500, step=50),
        'num_leaves':       trial.suggest_int('num_leaves', 15, 100),
        'max_depth':        trial.suggest_int('max_depth', 3, 10),
        'learning_rate':    trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'subsample':        trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'reg_alpha':        trial.suggest_float('reg_alpha', 1e-3, 10, log=True),
        'reg_lambda':       trial.suggest_float('reg_lambda', 1e-3, 10, log=True),
        'class_weight': 'balanced',
        'random_state': 42,
        'verbose': -1,
    }

    pipe = Pipeline([
        ('preproc', CreditPreprocessor()),
        ('encode', encoder),
        ('model', LGBMClassifier(**params)),
    ])

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring='average_precision')
    return scores.mean()

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

print("Best PR-AUC (CV):", study.best_value)
print("Best params:", study.best_params)
```

```python
## Пайплайн бустинга с лучшими параметрами
best_pipe = Pipeline([
    ('preproc', CreditPreprocessor()),
    ('encode', encoder),
    ('model', LGBMClassifier(**study.best_params, class_weight='balanced',
                             random_state=42, verbose=-1)),
])
best_pipe.fit(X_train, y_train)
y_proba_lgb = best_pipe.predict_proba(X_test)[:, 1]
print("LightGBM PR-AUC (test):", average_precision_score(y_test, y_proba_lgb))
print("LightGBM ROC-AUC (test):", roc_auc_score(y_test, y_proba_lgb))
```

Результаты оказались хуже, скорее всего в данных сильные линейные зависимости из-за которых логрег показыввает себя лучше

## Подбор гиперпараметров с Логрегом

```python
import optuna
from sklearn.model_selection import cross_val_score, StratifiedKFold


def objective_lr(trial):
    params = {
        'C':           trial.suggest_float('C', 1e-3, 100, log=True),
        'penalty':     trial.suggest_categorical('penalty', ['l1', 'l2']),
        'solver':      'liblinear', 
        'class_weight': 'balanced',
        'max_iter':    1000,
        'random_state': 42,
    }

    pipe = Pipeline([
        ('preproc', CreditPreprocessor()),
        ('encode', encoder),
        ('model', LogisticRegression(**params)),
    ])

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring='average_precision')
    return scores.mean()

study_lr = optuna.create_study(direction='maximize')
study_lr.optimize(objective_lr, n_trials=140)

print("Best PR-AUC (CV):", study_lr.best_value)
print("Best params:", study_lr.best_params)
```

```python
from sklearn.metrics import classification_report, roc_auc_score, average_precision_score

best_pipe_logreg = Pipeline([
    ('preproc', CreditPreprocessor()),
    ('encode', encoder),
    ('model', LogisticRegression(**study_lr.best_params, solver='liblinear',
                                 class_weight='balanced', max_iter=1000, random_state=42)),
])
best_pipe_logreg.fit(X_train, y_train)
y_proba_logreg = best_pipe_logreg.predict_proba(X_test)[:, 1]
print("Tuned LogReg PR-AUC (test):", average_precision_score(y_test, y_proba_logreg))
print("Tuned LogReg ROC-AUC (test):", roc_auc_score(y_test, y_proba_logreg))
```

- Подбор параметров не дал сильного увеличения (сравнимо с шумом)
- Признаков мало и они простые. ~6 базовых признаков (возраст, доход, история, сумма, срок, занятость). Это немного. Реальные скоринговые системы используют сотни признаков (поведенческие, транзакционные, внешние бюро). С 6 признаками 0.62 это честный максимум того, что можно выжать.
- Зависимости линейные. Логрег взял всё, что мог, бустинг не добавил. Значит сложных паттернов, которые подняли бы метрику, в данных просто нет.

## SHAP анализ результатов

```python
import shap
import numpy as np
import matplotlib.pyplot as plt

X_train_enc = best_pipe_logreg[:-1].transform(X_train)
X_test_enc  = best_pipe_logreg[:-1].transform(X_test)

# имена признаков после OneHotEncoder
feature_names = best_pipe_logreg.named_steps['encode'].get_feature_names_out()

# сама модель
model = best_pipe_logreg.named_steps['model']

#Создаём explainer для линейной модели 
explainer = shap.LinearExplainer(model, X_train_enc)
shap_values = explainer.shap_values(X_test_enc)

#Глобальная картина: какие признаки гонят дефолт
shap.summary_plot(
    shap_values, X_test_enc,
    feature_names=feature_names,
    show=False
)
plt.title("SHAP: влияние признаков на риск дефолта")
plt.tight_layout()
plt.savefig("shap_summary.png", dpi=120, bbox_inches='tight')
plt.show()

#Только важность (bar plot) усреднённый SHAP по признакам
shap.summary_plot(
    shap_values, X_test_enc,
    feature_names=feature_names,
    plot_type="bar",
    show=False
)
plt.title("SHAP: средняя важность признаков")
plt.tight_layout()
plt.savefig("shap_importance.png", dpi=120, bbox_inches='tight')
plt.show()
```

- **Выводы по SHAP-анализу**
-  **Топ-драйверы риска дефолта.** Модель опирается прежде всего на четыре признака: credit_history_bad (доминирующий фактор с большим отрывом), income, prev_loans и age. Направление влияния логично и согласуется со здравым смыслом: плохая кредитная история и низкий доход повышают риск, большое число прошлых займов также сдвигает прогноз в сторону дефолта.
- **Согласованность с EDA.** Главные драйверы по SHAP совпадают с закономерностями, выявленными на этапе разведочного анализа — модель не выучила артефактов, а уловила те же зависимости, что видны в данных. Отдельно стоит отметить заметный вклад географии (region: Москва, СПб) и возраста — эти факторы оказались значимее, чем можно было ожидать, что указывает на наличие региональной и возрастной структуры риска в данных.
- **Практический итог.** SHAP подтвердил интерпретируемость и осмысленность модели: решения объясняются понятными факторами кредитного риска, что критично для скоринга, где необходимо обосновывать отказы. Согласованность метрики, SHAP-значений и важности признаков повышает доверие к выводу о том, что модель достигла потолка предсказуемости на доступном наборе признаков.
